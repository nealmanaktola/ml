```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import tree
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from datetime import datetime
import warnings
from sklearn.model_selection import validation_curve, learning_curve
from utils import load_heart_data
import time
```

## Setup

```python
X, y = load_heart_data()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=23)
```

## Base Accuracy

This section will provide the base performance without any specific tuning

```python
nn = MLPClassifier(hidden_layer_sizes=(1,), random_state=7, max_iter=1000)
nn.fit(X_train, y_train)
y_pred = nn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: %.2f%%' % (accuracy * 100))
```

## Model Complexity Curve -  hidden_layer_sizes

```python
hidden_layer_range = np.linspace(1,150,30).astype('int')
hidden_layer_range = [(h, ) for h in hidden_layer_range]
train_scores, test_scores = validation_curve(nn, X_train, y_train, param_name="hidden_layer_sizes", param_range=hidden_layer_range, cv=5)
train_scores = np.mean(train_scores, axis=1)
test_scores = np.mean(test_scores, axis=1)
plt.figure()
plt.plot(hidden_layer_range, train_scores, label='Training score', color='r')
plt.plot(hidden_layer_range, test_scores, label='Cross Validation score', color='b')
plt.title('Model Complexity Curve: Hyperparameter: Hidden Layer Sizes (Hearts)')
#plt.xticks(hidden_layer_range)
plt.xlabel('Hidden Layer Sizes')
plt.ylabel('Score')
plt.tight_layout()
plt.grid()
plt.legend(loc="best")
plt.show()
```

## Model Complexity Curve - learning_rate_init

```python
lr_range = np.logspace(-5, 0, 6)
```

```python
train_scores, test_scores = validation_curve(nn, X_train, y_train, param_name="learning_rate_init", param_range=lr_range, cv=5)
train_scores = np.mean(train_scores, axis=1)
test_scores = np.mean(test_scores, axis=1)
plt.figure()
plt.semilogx(lr_range, train_scores, label='Training score', color='r')
plt.semilogx(lr_range, test_scores, label='Cross Validation score', color='b')
plt.title('Model Complexity Curve: Hyperparameter: Learning rate (Hearts)')
plt.xticks(lr_range)
plt.xlabel('Learning Rate')
plt.ylabel('Score')
plt.grid()
plt.legend(loc="best")
plt.show()
```

## Finding Hyperparameters

```python
lr_range = [0.001, 0.01, 0.05, 0.1]
hidden_layer_range = [1, 5, 10, 20, 40, 60, 80, 100]

param_grid = {'hidden_layer_sizes': hidden_layer_range, 'learning_rate_init': lr_range}
tree = GridSearchCV(nn, param_grid=param_grid, cv=5)
t0 = time.time()
tree.fit(X_train, y_train)
t1 = time.time()

train_time = t1 - t0
best_dt_params = tree.best_params_

t0 = time.time()
y_pred = tree.predict(X_test)
t1 = time.time()
test_time = t1 - t0
best_accuracy = accuracy_score(y_test, y_pred)

print('Inference time on test data: %f seconds' % test_time)
print('Accuracy is %.2f%%' % (best_accuracy * 100))
print('Completed training in %f seconds' % train_time)
print("Best parameters:")
print(best_dt_params)

```

## Learning Curve

```python
train_sizes = np.linspace(0.1, 1.0, 5)
_, train_scores, test_scores = learning_curve(tree, X_train, y_train, train_sizes=train_sizes, cv=5)

train_scores = np.mean(train_scores, axis=1)
test_scores = np.mean(test_scores, axis=1)
plt.figure()
plt.plot(train_sizes, train_scores, 'ro-', label='Training score')
plt.plot(train_sizes, test_scores, 'bo-', label='Cross Validation score')
plt.title('Learning curve Neural Network (Grades)')
plt.xlabel('Fraction of training examples')
plt.ylabel("Score")
plt.legend(loc="best")
plt.grid()
plt.show()
```

```python
nn = MLPClassifier(hidden_layer_sizes=(5,2), random_state=7, max_iter=1, warm_start=True)
nn.set_params(hidden_layer_sizes=best_dt_params['hidden_layer_sizes'], learning_rate_init=best_dt_params['learning_rate_init'])

iterations = 1000
train_loss = np.empty(iterations)
train_scores = np.empty(iterations)
val_scores = np.empty(iterations)

X_train1, X_val, y_train1, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=11)
for i in range(iterations):
    nn.fit(X_train1, y_train1)
    train_loss[i] = nn.loss_
    train_scores[i] = accuracy_score(y_train1, nn.predict(X_train1))
    val_scores[i] = accuracy_score(y_val, nn.predict(X_val))

y_pred = nn.predict(X_test)
nn_accuracy = accuracy_score(y_test, y_pred)
print('Accuracy of neural network is %.2f%%' % (nn_accuracy * 100))

xrange = np.arange(iterations) + 1
plt.figure()
plt.plot(xrange, train_loss)
plt.title('Training loss curve for neural network')
plt.xlabel('Epochs')
plt.ylabel("Loss")
plt.grid()
#plt.savefig(fig_path + 'nn_train_loss.png')
plt.show()

plt.figure()
plt.plot(xrange, train_scores, label='Training score')
plt.plot(xrange, val_scores, label='Validation score')
plt.title('Training and validation score curve for neural network')
plt.xlabel('Iterations')
plt.ylabel("Classification score")
plt.grid()
plt.legend(loc="best")
#plt.savefig(fig_path + 'nn_score_curve.png')
plt.show()

                                                         
                                            
```
